# Plotly.js based plots

The API of `Plotly.newPlot` is unchanged, except for the first parameter.

To use this module, you only need to run `npm i tslab-plotly` inside the folder where the notebook lives.

Visit https://github.com/dbuezas/tslab-plotly to report issues.

Visit https://plotly.com/javascript/ for more plotly.js examples and documentation.

# Example 1
A very simple plot

```typescript
import Plotly from "tslab-plotly";
import * as tslab from "tslab";

Plotly.newPlot(tslab, [{
  x: [1, 2, 3, 4, 5],
  y: [1, 2, 4, 8, 16],
}]);
```

# Example 2
3D Plots

```typescript
import Plotly from "tslab-plotly";
import * as tslab from "tslab";

const zs = [];
for (let x = -Math.PI; x < Math.PI; x += 0.1) {
  const row = [];
  zs.push(row);
  for (let y = -Math.PI; y < Math.PI; y += 0.1) {
    const z = Math.sin(x) * Math.cos(y);
    row.push(z);
  }
}
Plotly.newPlot(
  tslab,
  [
    {
      z: zs,
      type: "surface",
    },
  ],
  {
    title: "sin(x) * cos(x)",
    width: 500,
    height: 400,
  }
);
```

# Example 3
Contour plots

```typescript
import Plotly from "tslab-plotly";
import * as tslab from "tslab";

var layout = Plotly.newPlot(
  tslab,
  [
    {
      z: [
        [10, 10.625, 12.5, 15.625, 20],
        [5.625, 6.25, 8.125, 11.25, 15.625],
        [2.5, 3.125, 5.0, 8.125, 12.5],
        [0.625, 1.25, 3.125, 6.25, 10.625],
        [0, 0.625, 2.5, 5.625, 10],
      ],
      type: "contour",
    },
  ],
  {
    title: "Basic Contour Plot",
  }
);
```
